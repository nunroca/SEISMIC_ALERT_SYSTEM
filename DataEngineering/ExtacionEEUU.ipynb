```python
import numpy as np
from datetime import datetime,time,timedelta
```

```python
fecha_inicial = datetime(2018, 5, 3)
fecha_final = datetime(2023, 5, 3)
#intervalo de tiempo
diferencia_anual = timedelta(days=183)
# Generar secuencia de fechas
fechas = np.arange(fecha_inicial, fecha_final, diferencia_anual)
# Agregar fechas faltantes
if fechas[-1] < fecha_final:
    fechas = np.append(fechas,fecha_final)
#Convertir fechas a string
Fecha_sin_hora=[fecha.strftime("%Y-%m-%d") for fecha in  fechas]
print(Fecha_sin_hora)
```

```python
#creando funcion para ver las fechas y la cantidad de registros
def count_regsitro(fechas):
    for i in range(1,len(fechas)):
        url=f'https://earthquake.usgs.gov/fdsnws/event/1/count?format=geojson&starttime={fechas[i-1]}&endtime={fechas[i]}&minmagnitude=3'
        print('fecha: {} -{} registros: {}'.format(fechas[i-1],fechas[i],requests.get(url).json()['count']))
```

```python
count_regsitro(Fecha_sin_hora)
```

```python
#Creando funcion para extraer los datos de la api
def extraer_datos(fechas):
    data_df=pd.DataFrame()
    for i in range(1,len(fechas)):
        url=f'https://earthquake.usgs.gov/fdsnws/event/1/query?format=geojson&starttime={fechas[i-1]}&endtime={fechas[i]}&minmagnitude=3'
        data=requests.get(url).json()
        df=pd.json_normalize(data['features'])
        data_df=pd.concat([data_df,df],ignore_index=True)
    return data_df
```
